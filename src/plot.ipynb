```python
import pickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
with open('pred.pkl', 'rb') as f:
    data = pickle.load(f)
    
names, true, pred = data['name'], data['true'], data['pred']
```

```python
names_set = list(set(names))
num_names = len(names_set)

data = {name: {'area': [[], []], 'delay': [[], []]} for name in names_set}

for i, name in enumerate(names):
    for x in ['area', 'delay']:
        t, p = true[x][i], pred[x][i]
        data[name][x][0].append(t)
        data[name][x][1].append(p)
```

```python
colors = [cm.viridis(i / num_names) for i in range(num_names)]
name_to_color = {name: colors[i] for i, name in enumerate(names_set)}
```

```python
for i, name in enumerate(names_set):
    if name in ['sqrt', 'div']: continue
    plt.scatter(data[name]['area'][0], data[name]['area'][1], s=2, label=name)
plt.legend()
plt.title('Area without sqrt and div')
plt.xlabel('True area')
plt.ylabel('Predicted area')
plt.show()

for i, name in enumerate(names_set):
    plt.scatter(data[name]['area'][0], data[name]['area'][1], s=2, label=name)
plt.legend()
plt.title('Area all circuits')
plt.xlabel('True area')
plt.ylabel('Predicted area')
plt.show()
```

```python
for i, name in enumerate(names_set):
    if name in ['sqrt', 'div']: continue
    plt.scatter(data[name]['delay'][0], data[name]['delay'][1], s=2, label=name)
plt.legend()
plt.title('Delay without sqrt and div')
plt.xlabel('True delay')
plt.ylabel('Predicted delay')
plt.show()

for i, name in enumerate(names_set):
    plt.scatter(data[name]['delay'][0], data[name]['delay'][1], s=2, label=name)
plt.legend()
plt.title('Delay all circuits')
plt.xlabel('True delay')
plt.ylabel('Predicted delay')
plt.show()
```
